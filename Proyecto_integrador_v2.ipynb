# **Top deportistas mejor pagados en 1990-2020**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
global str
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Forbes Richest Atheletes (Forbes Richest Athletes 1990-2020).csv')
df.head(10)
```

```python
#Funcion cambio nombre deporte(unificarlos)
def cambio_deporte(sport):
    
    sport_dict = {
        'NASCAR': 'Auto Racing',
        'F1 Motorsports': 'Auto Racing',
        'auto racing': 'Auto Racing',
        'Auto Racing (Nascar)': 'Auto Racing',
        'Basketball': 'NBA',
        'basketball': 'NBA',
        'golf': 'Golf',
        'boxing': 'Boxing',
        'American Football': 'NFL',
        'soccer': 'Soccer',
        'F1 racing': 'Auto Racing',
        'tennis': 'Tennis',
        'motorcycle gp': 'Auto Racing',
        'baseball': 'Baseball',
        'Ice Hockey': 'Hockey',
        'cycling': 'Cycling',
        'ice hockey': 'Hockey',
        'Auto racing': 'Auto Racing',
        'Boxing': 'Boxing',
        'Golf': 'Golf',
        'Soccer': 'Soccer',
        'Tennis': 'Tennis',
        'Auto Racing': 'Auto Racing',
        'NFL': 'NFL',
        'Baseball': 'Baseball',
        'American Football / Baseball': 'American Football / Baseball',
        'Hockey': 'Hockey',
        'NBA': 'NBA',
        'MMA': 'MMA'
    }
    
    try:
        return sport_dict[sport]
    except KeyError:
        return '?'
```

```python
df.Sport=df.Sport.apply(cambio_deporte)
```

```python
df['Previous Year Rank']=df['Previous Year Rank'].fillna('not ranked')
```

```python
df.info()
```

Deportistas cuyo ranking ha subido al menos dos lugares entre 2010 y 2020.

```python
df_1020=df[df['Year']>2010]
```

```python
df_1020a=df_1020[['Name','Year','Current Rank']]
df_1020apvt=df_1020a.pivot_table(index='Name',columns='Year',values='Current Rank')
df_1020apvt
```

```python
sns.set(rc={'figure.figsize':(20,15)})
sns.heatmap(df_1020apvt)
plt.title('Apariciones en el ranking por Atleta')
```

Atleta con el menor nmero de apariciones y mayores ganancias.

```python
count_name=df['Name'].value_counts().to_frame()
count_name=count_name.reset_index(drop=False)
count_name=count_name.rename(columns={'index':'Name','Name':'Count name'})
count_name_1=count_name[count_name['Count name']==1]
earning_counts=pd.merge(count_name_1,df,how="left",on=["Name"])
Top_paid=earning_counts[earning_counts['earnings ($ million)']==earning_counts['earnings ($ million)'].max()]
Top_paid
```

```python
#Grafica conteo menores apariciones y ganancias
earning_counts_sort=earning_counts.sort_values("earnings ($ million)",ascending=True)
plt.figure(figsize=(10,8))
plt.barh(earning_counts_sort['Name'],earning_counts_sort['earnings ($ million)'])
plt.title('Atletas con menores apariciones y mayores ganancias')
plt.show()
```

Ganancia mnima y mxima (dentro del dataset) por deporte y por pas.

Ganancia mxima por deporte

```python
df_deporte=df[['Sport','earnings ($ million)']]
df_deporte_max=df_deporte.pivot_table(columns='Sport',values='earnings ($ million)',aggfunc=[np.max])
df_deporte_max=df_deporte_max.transpose().reset_index(drop=False)
df_deporte_max.drop(['level_0'], axis = 'columns', inplace=True)
df_deporte_max
```

Ganancia minima por deporte

```python
df_deporte_min=df_deporte.pivot_table(columns='Sport',values='earnings ($ million)',aggfunc=[np.min])
df_deporte_min=df_deporte_min.transpose().reset_index(drop=False)
df_deporte_min.drop(['level_0'], axis = 'columns', inplace=True)
df_deporte_min
```

```python
plt.figure(figsize=(6,4))
plt.plot('Sport','earnings ($ million)',data=df_deporte_max,marker="+",label="Ganancia mxima")
plt.plot('Sport','earnings ($ million)',data=df_deporte_min,marker="x",label="Ganancia minima")
plt.xticks(rotation=90)
plt.title('Ganancia minima y mxima por deporte')
plt.ylabel('earnings ($ million)')
plt.legend()
```

Ganancia mxima por nacionalidad

```python
df_Nationality=df[['Nationality','earnings ($ million)']]
df_Nationality_max=df_Nationality.pivot_table(columns='Nationality',values='earnings ($ million)',aggfunc=[np.max])
df_Nationality_max=df_Nationality_max.transpose().reset_index(drop=False)
df_Nationality_max.drop(['level_0'], axis = 'columns', inplace=True)
df_Nationality_max
```

Ganancia minima por nacionalidad

```python
df_Nationality_min=df_Nationality.pivot_table(columns='Nationality',values='earnings ($ million)',aggfunc=[np.min])
df_Nationality_min=df_Nationality_min.transpose().reset_index(drop=False)
df_Nationality_min.drop(['level_0'], axis = 'columns', inplace=True)
df_Nationality_min
```

```python
plt.figure(figsize=(6,4))
plt.plot('Nationality','earnings ($ million)',data=df_Nationality_max,marker="+",label="Ganancia mxima")
plt.plot('Nationality','earnings ($ million)',data=df_Nationality_min,marker="x",label="Ganancia minima")
plt.xticks(rotation=90)
plt.ylabel("earnings ($ million)")
plt.title("Ganancia mxima y minima por pais")
plt.legend()
```

Atleta con mayores ganancias por deporte por dcada

```python
decade90=df[df.Year.isin(np.arange(1990,1999))]
decade2000=df[df.Year.isin(np.arange(2000,2009,1))]
decade2010=df[df.Year.isin(np.arange(2010,2019,1))]
```

**Decada 90's (1990-1999)**

```python
decade90_=decade90.pivot_table(index=["Sport"],values=["earnings ($ million)"],aggfunc="max")
decade_90=pd.merge(decade90_,decade90,how="left",on=["Sport","earnings ($ million)"])
decade_90=decade_90.loc[:,["Name","Sport","earnings ($ million)"]]
decade_90
```

**Decada 2000 (2000-2009)**

```python
decade2000_=decade2000.pivot_table(index=["Sport"],values=["earnings ($ million)"],aggfunc="max")
decade_2000=pd.merge(decade2000_,decade2000,how="left",on=["Sport","earnings ($ million)"])
decade_2000=decade_2000.loc[:,["Name","Sport","earnings ($ million)"]]
decade_2000
```

**Decada 2010 (2010-2019)**

```python
decade2010_=decade2010.pivot_table(index=["Sport"],values=["earnings ($ million)"],aggfunc="max")
decade_2010=pd.merge(decade2010_,decade2010,how="left",on=["Sport","earnings ($ million)"])
decade_2010=decade_2010.loc[:,["Name","Sport","earnings ($ million)"]]
decade_2010
```

Ganancia total por cada deporte por cada ao

```python
df_dep_anio=df.pivot_table(index="Year",columns="Sport",values="earnings ($ million)",aggfunc="sum")
df_dep_anio.fillna(0)
df_dep_anio
```

```python
df_dep_anio.plot(kind="bar",stacked=True,figsize=(10,8))
plt.legend(bbox_to_anchor=(1.05, 1.0, 0.3, 0.2), loc='upper left')
plt.title("Ganancia por ao")
plt.ylabel("earnings ($ million)")
plt.grid(True,which='both')
```
